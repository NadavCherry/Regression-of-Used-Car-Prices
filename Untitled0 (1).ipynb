```python
# Extract horsepower
def extract_horsepower(engine_str):
    match = re.search(r'(\d+\.?\d*)\s*HP', engine_str)
    return float(match.group(1)) if match else None

# Extract engine volume
def extract_engine_volume(engine_str):
    match = re.search(r'(\d+\.\d+)\s*L', engine_str)
    return float(match.group(1)) if match else None

# Extract number of cylinders
def extract_cylinders(engine_str):
    # Cylinders
    cylinders_match = re.search(r'(\d+)\s*Cylinder|V(\d+)|Flat\s*(\d+)|Straight\s*(\d+)', engine_str)
    if cylinders_match:
        if cylinders_match.group(1):  # Case: '4 Cylinder'
            cylinders = int(cylinders_match.group(1))
        elif cylinders_match.group(2):  # Case: 'V6'
            cylinders = int(cylinders_match.group(2))
        elif cylinders_match.group(3):  # Case: 'Flat 6'
            cylinders = int(cylinders_match.group(3))
        elif cylinders_match.group(4):  # Case: 'Straight 6'
            cylinders = int(cylinders_match.group(4))
    else:
        cylinders = None

    return cylinders


# Extract fuel type
def extract_fuel_type(engine_str):
    if 'Gasoline' in engine_str:
        fuel_type = 'Gasoline'
    elif 'Diesel' in engine_str:
        fuel_type = 'Diesel'
    elif 'Electric' in engine_str:
        fuel_type = 'Electric'
    elif 'Plug-In Hybrid' in engine_str:
        fuel_type = 'Plug-In Hybrid'
    elif 'Hybrid' in engine_str:
        fuel_type = 'Hybrid'
    elif 'Flex Fuel' in engine_str:
        fuel_type = 'E85 Flex Fuel'
    elif engine_str == '–':
        fuel_type = 'Unknown'  # Handle missing data case
    elif engine_str == 'not supported':
        fuel_type = 'Unsupported'  # Handle unsupported fuel type case
    else:
        fuel_type = 'Unknown'
    return fuel_type

# Extract engine type (Turbo, TSI, TFSI, etc.)
def extract_engine_type(engine_str):
    engine_type = []

    if 'Turbo' in engine_str:
        engine_type.append('Turbo')
    if 'TSI' in engine_str:
        engine_type.append('TSI')
    if 'TFSI' in engine_str:
        engine_type.append('TFSI')
    if 'Supercharged' in engine_str:
        engine_type.append('Supercharged')
    if 'VTEC' in engine_str:
        engine_type.append('VTEC')
    if 'Twin Turbo' in engine_str:
        engine_type.append('Twin Turbo')

    return ', '.join(engine_type) if engine_type else 'Standard'

```

```python
# Apply the functions to the engine column
import re
df_train_data['horsepower'] = df_train_data['engine'].apply(extract_horsepower)
df_train_data['engine_volume'] = df_train_data['engine'].apply(extract_engine_volume)
df_train_data['cylinders'] = df_train_data['engine'].apply(extract_cylinders)
df_train_data['fuel_engine_type'] = df_train_data['engine'].apply(extract_fuel_type)
df_train_data['engine_type'] = df_train_data['engine'].apply(extract_engine_type)

# Apply the extraction functions to df_val_data
df_val_data['horsepower'] = df_val_data['engine'].apply(extract_horsepower)
df_val_data['engine_volume'] = df_val_data['engine'].apply(extract_engine_volume)
df_val_data['cylinders'] = df_val_data['engine'].apply(extract_cylinders)
df_val_data['fuel_engine_type'] = df_val_data['engine'].apply(extract_fuel_type)
df_val_data['engine_type'] = df_val_data['engine'].apply(extract_engine_type)

# Apply the extraction functions to test_data
test_data['horsepower'] = test_data['engine'].apply(extract_horsepower)
test_data['engine_volume'] = test_data['engine'].apply(extract_engine_volume)
test_data['cylinders'] = test_data['engine'].apply(extract_cylinders)
test_data['fuel_engine_type'] = test_data['engine'].apply(extract_fuel_type)
test_data['engine_type'] = test_data['engine'].apply(extract_engine_type)
```

```python
# Step 1: Extract the number of gears
def extract_gears(transmission_str):
    match = re.search(r'(\d+)-Speed', transmission_str)
    return int(match.group(1)) if match else 6.0

# Step 2: Determine if the transmission is automatic, manual, or CVT
def extract_transmission_type(transmission_str):
    transmission_str = transmission_str.lower()
    if 'a/t' in transmission_str or 'automatic' in transmission_str:
        return 'Automatic'
    elif 'm/t' in transmission_str or 'manual' in transmission_str:
        return 'Manual'
    elif 'cvt' in transmission_str:
        return 'CVT'
    else:
        return 'Unknown'

# Step 3: Identify additional flags (e.g., Auto-Shift, Overdrive, Dual Shift Mode)
def extract_additional_features(transmission_str):
    features = []
    if 'dual shift mode' in transmission_str.lower():
        features.append('Dual Shift Mode')
    if 'auto-shift' in transmission_str.lower():
        features.append('Auto-Shift')
    if 'overdrive' in transmission_str.lower():
        features.append('Overdrive')
    if 'dct' in transmission_str.lower():
        features.append('DCT')
    if 'fixed gear' in transmission_str.lower():
        features.append('Fixed Gear')

    return ', '.join(features) if features else 'None'
```

```python
# Step 4: Apply the functions to the transmission column for training, validation, and test data
for df in [df_train_data, df_val_data, test_data]:
    df['gears'] = df['transmission'].apply(extract_gears)
    df['transmission_type'] = df['transmission'].apply(extract_transmission_type)
    df['additional_features'] = df['transmission'].apply(extract_additional_features)

# Step 5: Check the new columns
print(df_train_data[['transmission', 'gears', 'transmission_type', 'additional_features']].head())
```

```python
def prepare_features(df, fields, is_tree_based=False):
    """
    Prepares a DataFrame with relevant features for regression or tree-based tasks.

    Parameters:
    df (pd.DataFrame): The original DataFrame.
    fields (list): A list of field names (features) to include.
    is_tree_based (bool): Whether the task is tree-based (True) or regression-based (False).
    
    Returns:
    pd.DataFrame: A new DataFrame with the relevant features.
    """

    # Initialize a new DataFrame with only the relevant fields
    df_filtered = df[fields].copy()

    # If it's a tree-based model (like Random Forest), handle categorical features with Label Encoding
    if is_tree_based:
        for field in fields:
            if df_filtered[field].dtype == 'object':
                df_filtered[field] = df_filtered[field].astype('category').cat.codes

    # If it's a regression-based model, handle categorical features with One-Hot Encoding
    else:
        df_filtered = pd.get_dummies(df_filtered, columns=[col for col in fields if df_filtered[col].dtype == 'object'], drop_first=True)

    return df_filtered
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, HistGradientBoostingRegressor
import xgboost as xgb
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.impute import SimpleImputer

# Load your dataset (assuming df_train_data, df_val_data, and test_data are ready)

# Prepare features and target
X_train = df_train_data[tree_features]  # Use relevant tree features
y_train = df_train_data['log_price']  # Target variable (you can use 'price' as well)

X_val = df_val_data[tree_features]
y_val = df_val_data['log_price']

# Step 1: Handle missing values for tree models (if needed)
imputer = SimpleImputer(strategy='mean')  # Using mean imputation for simplicity
X_train_imputed = imputer.fit_transform(X_train)
X_val_imputed = imputer.transform(X_val)

# Step 2: Define tree-based models to evaluate
tree_models = {
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(n_estimators=100, random_state=42),
    "Hist Gradient Boosting": HistGradientBoostingRegressor(max_iter=100, random_state=42),
    "XGBoost": xgb.XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=6, random_state=42)
}

# Step 3: Train models and evaluate
for name, model in tree_models.items():
    # Train the model with imputed (unscaled) data
    model.fit(X_train_imputed, y_train)
    
    # Predict on validation data
    y_pred = model.predict(X_val_imputed)

    # Evaluate performance
    mse = mean_squared_error(y_val, y_pred)
    r2 = r2_score(y_val, y_pred)
    
    print(f"{name} - MSE: {mse:.4f}, R^2: {r2:.4f}")
```
